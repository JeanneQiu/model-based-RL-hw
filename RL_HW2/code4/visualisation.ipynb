```python
import matplotlib.pyplot as plt

# Data from the experiment
n_values = [0, 1, 3, 5, 10, 20, 50, 100, 200, 300, 400, 500]
steps = [10600, 9000, 9200, 7400, 8200, 5800, 4800, 4300, 3300, 3000, 2700, 3000]
convergence_times = [16, 16, 14, 12, 15, 10, 9, 9, 7, 7, 7, 8]  # Converted to seconds

fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot convergence times on the left y-axis
ax1.set_xlabel('n (Number of Simulated Steps)', fontsize=12)
ax1.set_ylabel('Convergence Time (seconds)', color='blue', fontsize=12)
ax1.plot(n_values, convergence_times, marker='s', label='Convergence Time', color='blue', linestyle='-')
ax1.tick_params(axis='y', labelcolor='blue')
ax1.grid(True)

# Plot update steps on the right y-axis
ax2 = ax1.twinx()
ax2.set_ylabel('Total steps', color='green', fontsize=12)
ax2.plot(n_values, steps, marker='o', label='Update times', color='green', linestyle='--')
ax2.tick_params(axis='y', labelcolor='green')

# Add title and layout adjustments
# plt.title('Convergence Analysis: Time and Update times vs. n', fontsize=14)
fig.tight_layout()

# Show the plot
plt.show()
```
